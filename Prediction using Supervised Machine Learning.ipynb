### _Import the required modules_

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

### _Read the dataset_

```python
data_url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(data_url)
data.head(5)
```

### _Visualization_

```python
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

### _Preparing the data_

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

```python
X
```

```python
y
```

### _Model Training_

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

### **_Training the Algorithm_**
We have split our data into training and testing sets, and now is finally the time to train our algorithm. 

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
line = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

### **_Making Predictions_**
Now that we have trained our algorithm, it's time to make some predictions.

```python
y_pred = regressor.predict(X_test)
y_pred
```
